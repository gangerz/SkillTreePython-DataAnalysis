### 일변수 스칼라 함수
---

\begin{align}

\end{align}

#### **예제 01**

\begin{align}
f(x) = x^2+2x+1
\end{align}

* 함수 생성하기

````python
def f(x):
   return x**2 +2*x+1
#x=0
f(0)
````

* 인풋 정의 (-10~10, 100개)

````python
xs = np.linspace(-10, 10, 100)
f(xs)
````

````python
#plt.plot(x축, y축)
plt.plot(xs,f(xs), marker = '.')#''안에 들어가는 기호를 그래프로 표시
plt.xlabel('x')#x축 이름
plt.ylabel('y')#y축 이름
````

* 그래프 설명

### 다변수 스칼라 함수
---

여러개의 인풋(=다변수), 하나의 아웃풋(=스칼라, 스칼라함수)


\begin{align}
y = f(𝐱)\\
\end{align}



#### **예제 03**

\begin{align}
f(x, y) = x+y
\end{align}

* 함수 생성하기

````python
def f3(x,y):
  return x+y
  
````

````python
f3(1,5)
````

* 인풋 정의 (x, y, 모두 -15~20 값)

````python
xs = np.linspace(-15,20,12)
ys = np.linspace(-15,20,12)
X, Y = np.meshgrid(xs,ys)#모든 순서 쌍 만듬
print(X,Y)
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize=(9,7)) # 그림판
ax = plt.axes(projection='3d') #축

ax.scatter3D(X, Y, f3(X,Y), marker ='.', c='black')#점을 찍는것
surf=ax.plot_surface(X,Y,f3(X,Y), alpha=0.6, cmap='rainbow')#면  
fig.colorbar(surf, shrink=0.6) #surf를 기반으로 데이터 막대 표시
#surface : 면으로 나타냄
#alpha : 진하기
#cmap : 면 색깔
#c : 점 색깔
#shrink : 막대 크기 조절
#기타 제목/축제목 등등...
ax. set_title('f(x,y)') #그래프 제목
ax. set_xlabel('x')    #그래프 X축
ax. set_ylabel('y')    #그래프 Y축
ax. set_zlabel('f(x,y)')#그래프 Z축
plt.show()
````

* 그래프 설명
  * x, y가 가질수 있는 값 (인풋) : 
  * f(x)가 가질 수 있는 값 (아웃풋) : 
  * 그래프의 특징 :

## 일변수 벡터함수
---

하나의 인풋(=스칼라, 일변수), 여러개의 아웃풋(=스칼라, 스칼라함수)




\begin{align}
f(x) = (f_1(x), f_2(x), ..., f_n(x))^T
\end{align}

#### **예제 06**

\begin{align}
f(x) = \begin{pmatrix}
h(x)\\ 
g(x)\\ 
\end{pmatrix}  = \begin{pmatrix}
x^2\\ 
5x\\ 
\end{pmatrix}
\end{align}

* 함수 $f(x)$ 생성하기

````python
def f6(x):
 h = x**2
 g = 5*x
 return np.array([h, g])

h, g = f6(1)
print(h,g)

````

* 인풋 정의 (-10~10 20개)

````python
xs = np.linspace(-10,10,21)
hs, gs = f6(xs)
print(xs,h,g)
print(hs)
print(gs)
````

* 2d플롯으로 시각화

````python
#인풋은 색깔로 표현
# 아웃풋은 2개를 2차원에 그리기
plt.plot(gs, hs) # plot은 x값과 y값을 순서대로 적용
scat = plt.scatter(gs,hs,c=xs, cmap='rainbow') # 점그리기  , c=xs으로 x 값에 따른 색깔 표시
plt.colorbar(scat)
plt.xlabel('g(x)')
plt.ylabel('h(x)')
````

* 그래프 설명
  * x가 가질수 있는 값 (인풋) : 
  * g(x)가 가질 수 있는 값 (첫번째 아웃풋) : 
  * h(x)가 가질수 있는 값 (두번째 아웃풋): 

#### **예제 07**

\begin{align}
f(t) = \begin{pmatrix}
x(t)\\ 
y(t)\\ 
z(t)\\ \end{pmatrix} = \begin{pmatrix}
sin(6t)\\
\frac{t}{4}\\
\frac{t^2}{2}\\
\end{pmatrix}
\end{align}

* 함수 생성하기

````python
def f7(t):
  x = np.sin(6*t)
  y = t/4
  z = (t**2)/2
  return np.array([x,y,z])
f7(1)
````

* 인풋 정의 (-1~3 200개)

````python
ts = np.linspace(-1,3,200)
xs, ys, zs = f7(ts)
print(ts.shape, xs.shape, ys.shape, zs.shape)
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize=(9,7)) # 그림판
ax = plt.axes(projection='3d') #축

scat = ax.scatter3D(xs,ys,zs, marker ='.', c=ts)#점을 찍는것
#ax.plot_surface(xs,ys,zs, alpha=0.6, cmap='rainbow')#면  
fig.colorbar(scat, shrink=0.6) #surf를 기반으로 데이터 막대 표시
#surface : 면으로 나타냄
#alpha : 진하기
#cmap : 면 색깔
#c : 점 색깔
#shrink : 막대 크기 조절
#기타 제목/축제목 등등...
ax. set_title('$f(t)outputs') #그래프 제목
ax. set_xlabel('$sin(6t)')    #그래프 X축
ax. set_ylabel('$t/4')    #그래프 Y축
ax. set_zlabel('$t^2/2')#그래프 Z축
plt.show()
````

* 그래프 설명
  * t가 가질수 있는 값 (인풋) : 
  * x(t)가 가질 수 있는 값 (첫번째 아웃풋) : 
  * y(t)가 가질 수 있는 값 (두번째 아웃풋) : 
  * z(t)가 가질 수 있는 값 (세번째 아웃풋) : 

#### **예제 08**

\begin{align}

y(t)\\ 
z(t)\\ \end{pmatrix} = \begin{pmatrix}
cos(t)\\
sin(t)\\
t\\
\end{pmatrix}

* 함수 생성하기

* 그래프 설명
  * t가 가질수 있는 값 (인풋) : 
  * cos(t)가 가질 수 있는 값 (첫번째 아웃풋) : 
  * sing(t)가 가질 수 있는 값 (두번째 아웃풋) : 


## 다변수 벡터함수
---
여러개의 인풋(=스칼라, 일변수), 여러개의 아웃풋(=스칼라, 스칼라함수)

\begin{align}
f(𝐱) = (f_1(𝐱), f_2(𝐱), ..., f_n(𝐱))^T
\end{align}

#### **예제 09**


\begin{align}
f(x,y) = \begin{pmatrix}
x^2\\ 
y^2\\ 
xy\\ \end{pmatrix}
\end{align}

* 함수 생성하기

````python
def f9(x,y):
  
  return np.array([x**2, y**2, x*y])

f9(1,2)
````

* 인풋 정의 (x,y 모두 -0.5 ~ 1 사이 50개)

````python
xs = np.linspace(-0.5,1,50)
ys = np.linspace(-0.5,1,50)
X, Y = np.meshgrid(xs,ys)#모든 순서 쌍 만듬
print(X,Y)
print(X.shape,Y.shape)
````

````python
F = f9(X,Y)
F.shape
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize=(9,7)) # 그림판
ax = plt.axes(projection='3d') #축

ax.scatter3D(F[0],F[1],F[2], marker ='.', c='black')#점을 찍는것
surf=ax.plot_surface(F[0],F[1],F[2], alpha=0.6, cmap='rainbow')#면  
#fig.colorbar(surf, shrink=0.6) #surf를 기반으로 데이터 막대 표시
#surface : 면으로 나타냄
#alpha : 진하기
#cmap : 면 색깔
#c : 점 색깔
#shrink : 막대 크기 조절
#기타 제목/축제목 등등...
ax. set_title('f(x,y)output') #그래프 제목
ax. set_xlabel('$x^2$')    #그래프 X축
ax. set_ylabel('$y^2$')    #그래프 Y축
ax. set_zlabel('$xy$')#그래프 Z축
plt.show()
````

#### **예제 10**


\begin{align}
f(u,v) = \begin{pmatrix}
u\\ 
v\\ 
1+u^2+\frac{v}{1+v^2}\\ \end{pmatrix}, \ \ \ \ \ 0 \leq u, v \leq1
\end{align}

* 함수 생성하기

````python
def f0(u,v):
  
  return np.array([u, v, 1+u**2+v/(1+v**2)])

f9(1,2)
````

* 인풋 정의 (u, v 모두 0 ~ 1 사이 50개)

````python
us = np.linspace(0,1,50)
vs = np.linspace(0,1,50)
U, V = np.meshgrid(us,vs)#모든 순서 쌍 만듬
print(U,V)
print(U.shape,V.shape)
````

````python
F = f0(U,V)
F.shape
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize=(9,7)) # 그림판
ax = plt.axes(projection='3d') #축

ax.scatter3D(F[0],F[1],F[2], marker ='.', c='black')#점을 찍는것
surf=ax.plot_surface(F[0],F[1],F[2], alpha=0.6, cmap='rainbow')#면  
#fig.colorbar(surf, shrink=0.6) #surf를 기반으로 데이터 막대 표시
#surface : 면으로 나타냄
#alpha : 진하기
#cmap : 면 색깔
#c : 점 색깔
#shrink : 막대 크기 조절
#기타 제목/축제목 등등...
ax. set_title('f(u,v)output') #그래프 제목
ax. set_xlabel('$u$')    #그래프 X축
ax. set_ylabel('$v$')    #그래프 Y축
ax. set_zlabel('$1+u**2+v/(1+v**2)$')#그래프 Z축
plt.show()
````
